```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering, KMeans
import matplotlib.pyplot as plt
from sklearn import manifold
from sklearn import preprocessing

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,5)
```

```python
X = np.array([
    [0.6, 1.9],
    [1.8, 1.6],
    [2.7, 2.0],
    [3.0, 2.1],
    [3.0, 2.6],
    [3.1, 4.5],
    [3.8, 0.6],
    [4.2, 2.7]
])
```

# Agglomerative clustering

## Single link

```python
model = AgglomerativeClustering(n_clusters=2, affinity='cosine', linkage='single')
model.fit(X)

plt.figure()
plt.scatter(X[model.labels_==0,0],X[model.labels_==0,1], c='r' )
plt.scatter(X[model.labels_==1,0],X[model.labels_==1,1], c='b' )
plt.show()
```

## Complete link

```python
model = AgglomerativeClustering(n_clusters=2, affinity='cosine', linkage='complete')
model.fit(X)

plt.figure()
plt.scatter(X[model.labels_==0,0],X[model.labels_==0,1], c='r' )
plt.scatter(X[model.labels_==1,0],X[model.labels_==1,1], c='b' )
plt.show()
```

# K-means 

Initial obejcts - e and f

```python
X_n = preprocessing.normalize(X, axis=1)
```

```python
init_obj = np.array([
    X_n[4], # e
    X_n[5]  # f
])

model = KMeans(n_clusters=2, init=init_obj, n_init=1, verbose=1)
model.fit(X)

plt.figure()
plt.scatter(X[model.labels_==0,0],X[model.labels_==0,1], c='r' )
plt.scatter(X[model.labels_==1,0],X[model.labels_==1,1], c='b' )
plt.show()
```

# Вывод

Кластеризация в KMeans выполнена за одну итерацию (на второй перемещения центроид уже не происходит), поскольку все точки, не являющиеся центроидами, расположены ближе к точке E. "Визуально" неоднозначно, какая из кластеризаций лучше, поскольку объекты расположены в форме "креста". На мой взгляд, лучшей кластеризацией была бы следующая (использованные методы не смогли бы так разделить в силу спейифика алгоритма)

```python
plt.figure()
plt.scatter([2.7, 3.0, 3.0], [2.0, 2.1, 2.6], c='r' )
plt.scatter([0.6, 1.8, 3.1, 3.8, 4.2], [1.9, 1.6, 4.5, 0.6, 2.7], c='b' )
plt.show()
```
